## 1.1 Vectors in Several Dimensions

As stated in the notes (§1.1), we use **boldface** for vectors: $\mathbf{u} \in \mathbb{R}^n$.  
In SymPy we represent vectors as `Matrix` objects.

```python
# Definition 1.1.1 — Vector addition and scalar multiplication
u = Matrix([0, 1, 3])
v = Matrix([7, -2, 10])

print('u + v =')
display(u + v)  # Expected: (7, -1, 13) — cf. Example 1.1 in the notes

# Scalar multiplication (Example 1.1, item 3)
a = Matrix([2, 0, sqrt(2)])
k_val = 7
print(f'\n{k_val} * a =')
display(k_val * a)  # Expected: (14, 0, 7√2)
```

```python
# Canonical base in R^3 (Definition 1.1.2)
i_hat = Matrix([1, 0, 0])
j_hat = Matrix([0, 1, 0])
k_hat = Matrix([0, 0, 1])

# Any vector (a, b, c) = a*i + b*j + c*k
a_val, b_val, c_val = 3, -1, 5
vec = a_val * i_hat + b_val * j_hat + c_val * k_hat
print('3i - j + 5k =')
display(vec)
```

## 1.2 Dot Product, Norm, Angle, Cross Product

From the notes:
- **Dot product** (Def. 1.1.3): $\mathbf{u} \cdot \mathbf{v} = u_1 v_1 + u_2 v_2 + \cdots + u_n v_n$
- **Norm** (Def. 1.1.4): $\|\mathbf{v}\| = \sqrt{v_1^2 + v_2^2 + \cdots + v_n^2}$
- **Angle** (Def. 1.1.5): $\cos\theta = \frac{\mathbf{u}\cdot\mathbf{w}}{\|\mathbf{u}\|\|\mathbf{w}\|}$

```python
u = Matrix([1, 2, 3])
w = Matrix([4, -1, 2])

# Dot product
dot_uw = u.dot(w)
print('u · w =', dot_uw)

# Norm
norm_u = u.norm()
norm_w = w.norm()
print('||u|| =', norm_u, '=', simplify(norm_u))
print('||w|| =', norm_w, '=', simplify(norm_w))

# Angle between u and w
cos_theta = dot_uw / (norm_u * norm_w)
angle = acos(cos_theta)
print('\nAngle θ (radians) =')
display(simplify(angle))

# Verify Cauchy-Schwarz inequality (Theorem 1.1.1): |u·v| ≤ ||u|| ||v||
print('\nCauchy-Schwarz check:', Abs(dot_uw), '≤', simplify(norm_u * norm_w),
      '→', bool(Abs(dot_uw) <= simplify(norm_u * norm_w)))
#
```

```python
# Cross product (Definition 1.1.6) — only defined in R^3
a_vec = Matrix([1, 0, 0])
b_vec = Matrix([0, 1, 0])

cross_ab = a_vec.cross(b_vec)
print('a × b =')
display(cross_ab)  # Should be (0, 0, 1) = k

# Verify orthogonality: (a × b) · a = 0 and (a × b) · b = 0
print('(a × b) · a =', cross_ab.dot(a_vec))
print('(a × b) · b =', cross_ab.dot(b_vec))

# Magnitude = area of parallelogram (||a|| ||b|| sin θ)
print('||a × b|| =', cross_ab.norm())
```

## 1.3 Distance, Projection, and Parametric Lines

- **Distance** (Def. 1.1.7): $d(\mathbf{P},\mathbf{Q}) = \|\mathbf{P} - \mathbf{Q}\|$
- **Orthogonal projection** (Def. 1.1.8): $\mathrm{proj}_{\mathbf{a}} \mathbf{v} = \frac{\mathbf{a} \cdot \mathbf{v}}{\|\mathbf{a}\|^2}\,\mathbf{a}$
- **Parametric line** (Def. 1.1.9): $\mathbf{r}(t) = \mathbf{b} + t\,\mathbf{a}$

```python
# Distance between two points
P = Matrix([1, 2, 3])
Q = Matrix([4, 6, 3])
d_PQ = (P - Q).norm()
print('d(P, Q) =', simplify(d_PQ))

# Orthogonal projection of v onto a
a_proj = Matrix([1, 1, 0])
v_proj = Matrix([3, 1, 2])
proj = (a_proj.dot(v_proj) / a_proj.dot(a_proj)) * a_proj
print('\nproj_a(v) =')
display(proj)
```

```python
# Parametric equation of a line (Example from notes §1.1)
# Line through (1, -2, 3) parallel to (π, -3, 1)
b_line = Matrix([1, -2, 3])
a_line = Matrix([pi, -3, 1])

r_line = b_line + t * a_line
print('r(t) =')
display(r_line)
# Parametric equations: x = πt + 1, y = -3t - 2, z = t + 3
```

## 1.4 Planes and Distance from a Point to a Plane

From the notes (§1.2):
- Plane: $Ax + By + Cz = D$, normal vector $\mathbf{n} = (A, B, C)$
- Distance from point $P_0(x_0,y_0,z_0)$ to plane: $d = \frac{|Ax_0 + By_0 + Cz_0 - D|}{\sqrt{A^2+B^2+C^2}}$

```python
# Example from notes: Plane 2x - 3y + 4z = 12, point P0 = (1, 2, 3)
A, B, C, D = 2, -3, 4, 12
x0, y0, z0 = 1, 2, 3

numerator = Abs(A*x0 + B*y0 + C*z0 - D)
denominator = sqrt(A**2 + B**2 + C**2)
distance = numerator / denominator

print('Distance from P0 to the plane =')
display(simplify(distance))  # Should be 4/√29
```

## 1.5 Coordinate Systems: Polar, Cylindrical, Spherical

As defined in the notes (§1.3):

| System | Coordinates | Cartesian relations |
|--------|-------------|--------------------|
| Polar | $(r,\theta)$ | $x = r\cos\theta,\; y = r\sin\theta$ |
| Cylindrical | $(r,\theta,z)$ | $x = r\cos\theta,\; y = r\sin\theta,\; z=z$ |
| Spherical | $(\rho,\theta,\phi)$ | $x = \rho\sin\phi\cos\theta,\; y=\rho\sin\phi\sin\theta,\; z=\rho\cos\phi$ |

```python
# Polar → Cartesian (Example from notes: (4, π/4) → Cartesian)
r_val, theta_val = 4, pi/4
x_cart = r_val * cos(theta_val)
y_cart = r_val * sin(theta_val)
print(f'Polar ({r_val}, π/4) → Cartesian:')
display(Matrix([simplify(x_cart), simplify(y_cart)]))  # (2√2, 2√2)

# Cartesian → Cylindrical (Example: (3, 3, 5))
x_c, y_c, z_c = 3, 3, 5
r_cyl = sqrt(x_c**2 + y_c**2)
theta_cyl = atan2(y_c, x_c)
print(f'\nCartesian ({x_c},{y_c},{z_c}) → Cylindrical:')
display(Matrix([simplify(r_cyl), simplify(theta_cyl), z_c]))  # (3√2, π/4, 5)

# Cartesian → Spherical (Example: (1, 1, √2))
x_s, y_s, z_s = 1, 1, sqrt(2)
rho_sph = sqrt(x_s**2 + y_s**2 + z_s**2)
theta_sph = atan2(y_s, x_s)
phi_sph = acos(z_s / rho_sph)
print(f'\nCartesian (1, 1, √2) → Spherical (ρ, θ, φ):')
display(Matrix([simplify(rho_sph), simplify(theta_sph), simplify(phi_sph)]))  # (2, π/4, π/4)
```

## 2.1 Functions of Several Variables — Domain, Range, Composition

```python
# Scalar function: f(x,y) = x^2 + y^2
f_scalar = x**2 + y**2
print('f(x,y) =')
display(f_scalar)

# Evaluate at a point
print('f(1,2) =', f_scalar.subs([(x, 1), (y, 2)])) #subs is substitution

# Vectorial function: f(x,y) = (x, y, x+y) — cf. Example in notes
f_vec = Matrix([x, y, x + y])
print('\nf(x,y) =')
display(f_vec)
```

## 2.2 Limits of Functions of Several Variables

As explained in the notes (§2.3), a limit in $\mathbb{R}^n$ exists only if the limit along **all** paths agrees. We can disprove existence by finding two paths with different limits.

```python
# Example from notes: f(x,y) = xy / (x^2 + y^2)
# Approach (0,0) along different paths

f_lim = x*y / (x**2 + y**2)

# Path 1: y = 0
path1 = f_lim.subs(y, 0)  # 0/x^2 = 0
lim1 = limit(path1, x, 0)
print('Along y=0:', lim1)

# Path 2: y = x
path2 = f_lim.subs(y, x)  # x^2/(2x^2) = 1/2
lim2 = limit(path2, x, 0)
print('Along y=x:', lim2)

print('\n→ The limits differ (0 ≠ 1/2), so the limit does NOT exist.')
```

```python
# Example from notes: show lim x^2*y / (x^2+y^2) = 0
# Using the bounded × infinitesimal argument
f_bounded = x**2 / (x**2 + y**2)

# Along any path y = m*x: f_bounded = x^2/(x^2+m^2*x^2) = 1/(1+m^2) ≤ 1
m = symbols('m', real=True)
f_along_mx = f_bounded.subs(y, m*x)
print('g(x,y) along y=mx =')
display(simplify(f_along_mx))  # 1/(1+m^2), which is bounded
print('\nSince g is bounded and y→0 is infinitesimal, the product → 0.')
```

## 3.1 Partial Derivatives and the Gradient

From the notes (§3.1):
- $\frac{\partial f}{\partial x}$: differentiate treating other variables as constants
- **Gradient**: $\nabla f = \left(\frac{\partial f}{\partial x_1}, \ldots, \frac{\partial f}{\partial x_n}\right)$

```python
# Example from notes: f(x,y) = x^2 + y^3
f1 = x**2 + y**3

df_dx = diff(f1, x) #nombres muy obvios. Por ejemplo, en este caso la derivada parcial de x es nombrada df_dx
df_dy = diff(f1, y)
print('∂f/∂x ='); display(df_dx)  # 2x
print('∂f/∂y ='); display(df_dy)  # 3y^2

# Example: f(x,y,z) = xyz (notes §3.1)
f2 = x * y * z
grad_f2 = Matrix([diff(f2, x), diff(f2, y), diff(f2, z)]) #gradiente en forma matricial
print('\n∇f(x,y,z) for f = xyz:')
display(grad_f2)  # (yz, xz, xy)
```

```python
# Helper function: compute the gradient of a scalar function
#cuando definamos un vector va a ser a column vector
def gradient(f, variables):
    """Compute ∇f as a column Matrix."""
    return Matrix([diff(f, var) for var in variables])

# Example from notes: f(x,y,z) = x^2 + y^2 + z^2
f3 = x**2 + y**2 + z**2
grad_f3 = gradient(f3, [x, y, z])
print('∇f =')
display(grad_f3)

# Evaluate at (1, 1, 1)
grad_at_111 = grad_f3.subs([(x, 1), (y, 1), (z, 1)])
print('∇f(1,1,1) =')
display(grad_at_111)  # (2, 2, 2)

# Magnitude of the gradient = rate of steepest ascent
print('||∇f(1,1,1)|| =', simplify(grad_at_111.norm()))  # 2√3
display(simplify(grad_at_111.norm()))
```

## 3.2 Tangent Plane

From the notes (§3.1, Tangent hyper-plane):
- For an explicit surface $z = f(x,y)$, the tangent plane at $(x_0,y_0)$ is:

$$z = f(x_0,y_0) + f_x(x_0,y_0)(x-x_0) + f_y(x_0,y_0)(y-y_0)$$

```python
# Tangent plane to f(x,y) = x^2 + y^2 at (1, 1)
f_tp = x**2 + y**2
x0_tp, y0_tp = 1, 1

f_val = f_tp.subs([(x, x0_tp), (y, y0_tp)])
fx_val = diff(f_tp, x).subs([(x, x0_tp), (y, y0_tp)])
fy_val = diff(f_tp, y).subs([(x, x0_tp), (y, y0_tp)])

tangent_plane = f_val + fx_val*(x - x0_tp) + fy_val*(y - y0_tp)
print('Tangent plane at (1,1): z =')
display(sp.expand(tangent_plane))  # 2 + 2(x-1) + 2(y-1) = 2x + 2y - 2
```

## 3.3 The Chain Rule

From the notes (§3.2): if $z = f(x,y)$ with $x=x(t)$ and $y=y(t)$, then

$$\frac{dz}{dt} = \frac{\partial z}{\partial x}\frac{dx}{dt} + \frac{\partial z}{\partial y}\frac{dy}{dt}$$

```python
# Example from notes: z = x^2 + y^2, x = t^2, y = sin(t)
z_expr = x**2 + y**2
x_of_t = t**2
y_of_t = sin(t)

# Chain rule: dz/dt = (∂z/∂x)(dx/dt) + (∂z/∂y)(dy/dt)
dz_dt_chain = (
    diff(z_expr, x) * diff(x_of_t, t) +
    diff(z_expr, y) * diff(y_of_t, t)
)

# Substitute x(t) and y(t)
dz_dt_chain = dz_dt_chain.subs([(x, x_of_t), (y, y_of_t)])
print('dz/dt (via chain rule) =')
display(simplify(dz_dt_chain))  # 4t^3 + 2sin(t)cos(t)

# Verify by direct substitution and differentiation
z_direct = z_expr.subs([(x, x_of_t), (y, y_of_t)])
dz_dt_direct = diff(z_direct, t)
print('dz/dt (direct) =')
display(simplify(dz_dt_direct))

print('\nBoth agree:', simplify(dz_dt_chain - dz_dt_direct) == 0)
```

## 3.4 Directional Derivatives

From the notes (§3.3): The directional derivative of $f$ at $\mathbf{a}$ in the direction of unit vector $\mathbf{u}$ is

$$D_{\mathbf{u}} f(\mathbf{a}) = \nabla f(\mathbf{a}) \cdot \mathbf{u}$$

```python
# Example from notes: f(x,y) = x^2 + y^2 at a = (1,1),
# direction u = (1/√2, 1/√2)

f_dd = x**2 + y**2
a_point = {x: 1, y: 1}
u_dir = Matrix([1/sqrt(2), 1/sqrt(2)])

grad_f_dd = gradient(f_dd, [x, y]).subs(a_point)
D_u = grad_f_dd.dot(u_dir)

print('∇f(1,1) ='); display(grad_f_dd)
print('D_u f(1,1) ='); display(simplify(D_u))  # 2√2

# The maximum rate of change is ||∇f|| in the direction of ∇f
print('\nMax rate of change = ||∇f(1,1)|| =', simplify(grad_f_dd.norm()))
```

## 3.5 Higher-Order Derivatives and the Hessian Matrix

From the notes (§3.4):
- **Clairaut's Theorem**: if mixed partials are continuous, $\frac{\partial^2 f}{\partial x\partial y} = \frac{\partial^2 f}{\partial y\partial x}$
- The **Hessian** $H_f$ collects all second-order partial derivatives

```python
# Helper: compute the Hessian matrix
def hessian(f, variables):
    """Compute the Hessian matrix H_f."""
    n = len(variables)
    H = sp.zeros(n, n)
    for i in range(n):
        for j in range(n):
            H[i, j] = diff(f, variables[i], variables[j])
    return H

# Example from notes: f(x,y,z) = x^2*y + y*z^3 + sin(x*z)
f_hess = x**2 * y + y * z**3 + sin(x * z)
H = hessian(f_hess, [x, y, z])
print('Hessian of f(x,y,z) = x²y + yz³ + sin(xz):')
display(H)

# Verify symmetry (Clairaut's theorem)
print('\nH is symmetric:', H.equals(H.T)) #Esta T significa que transforma la matriz en traspuesta
```

## 3.6 The Jacobian Matrix

From the notes (§3.5): for $\mathbf{f}: \mathbb{R}^n \to \mathbb{R}^m$, the Jacobian matrix is

$$Df(\mathbf{a}) = \begin{bmatrix} \frac{\partial f_1}{\partial x_1} & \cdots & \frac{\partial f_1}{\partial x_n} \\ \vdots & \ddots & \vdots \\ \frac{\partial f_m}{\partial x_1} & \cdots & \frac{\partial f_m}{\partial x_n} \end{bmatrix}$$

```python
# Example from notes: f(x,y) = (x^2 + y, xy)
f_jac = Matrix([x**2 + y, x*y])
variables = [x, y]

J = f_jac.jacobian(variables)
print('Jacobian of f(x,y) = (x²+y, xy):')
display(J)  # [[2x, 1], [y, x]]

# Chain rule in matrix form: D(f∘g) = Df(g(a)) · Dg(a)
# Example from notes: f(x,y)=(x^2,y^2), g(t)=(t, sin(t))
f_comp_mat = Matrix([x**2, y**2])
g_comp_mat = Matrix([t, sin(t)])

Df = f_comp_mat.jacobian([x, y])
Dg = g_comp_mat.jacobian([t])

# Evaluate Df at g(t)
Df_at_g = Df.subs([(x, t), (y, sin(t))])
D_composition = Df_at_g * Dg

print('\nD(f∘g)(t) = Df(g(t)) · Dg(t) =')
display(simplify(D_composition))  # [2t, 2sin(t)cos(t)]^T
```

## 4.1 Taylor Expansion in Several Variables

From the notes (§4.1): the second-order Taylor expansion of $f$ around $\mathbf{x}_0$ is

$$f(\mathbf{x}) \approx f(\mathbf{x}_0) + \nabla f(\mathbf{x}_0)\cdot(\mathbf{x}-\mathbf{x}_0) + \frac{1}{2}(\mathbf{x}-\mathbf{x}_0)^T H_f(\mathbf{x}_0)(\mathbf{x}-\mathbf{x}_0)$$

```python
# Example from notes: f(x,y) = sin(xy) at (1, π/2)
f_taylor = sin(x * y)
x0_t, y0_t = 1, pi/2
subs_point = [(x, x0_t), (y, y0_t)]

# f(x0, y0)
f0 = f_taylor.subs(subs_point)
print('f(1, π/2) ='); display(f0)  # 1

# Gradient at the point
grad_ft = gradient(f_taylor, [x, y])
grad_ft_val = grad_ft.subs(subs_point)
print('∇f(1,π/2) ='); display(grad_ft_val)  # (0, 0)

# Hessian at the point
H_ft = hessian(f_taylor, [x, y])
H_ft_val = H_ft.subs(subs_point)
print('H_f(1,π/2) ='); display(H_ft_val)  # [[-π²/4, -π/2], [-π/2, -1]]

# Build the second-order Taylor approximation
h_vec = Matrix([x - x0_t, y - y0_t])
taylor_2 = f0 + grad_ft_val.dot(h_vec) + Rational(1, 2) * h_vec.dot(H_ft_val * h_vec)
print('\nSecond-order Taylor approximation:')
display(sp.expand(taylor_2))
```

### Visualization: First- and Second-Order Taylor Approximations

We plot the original surface $f(x,y) = \sin(xy)$ together with:
- The **first-order (linear) approximation** — a tangent plane at $(1, \pi/2)$
- The **second-order (quadratic) approximation** — a paraboloid that captures curvature

This illustrates how higher-order Taylor terms improve the local fit (cf. Notes §4.1, Theorem on the Second-Order Taylor Formula).

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# --- Symbolic expressions (computed in the cell above) ---
# f(x,y) = sin(xy), expansion point (1, π/2)
x0_v, y0_v = 1.0, float(pi/2)

# First-order (linear) Taylor approximation:  l(x,y) = f(x0,y0) + ∇f·(x-x0,y-y0)
# From the notes example: f(1,π/2) = 1, f_x = 0, f_y = 0  →  l(x,y) = 1
taylor_1_expr = f0 + grad_ft_val.dot(h_vec)

# Second-order Taylor approximation (already computed as taylor_2)
# #taylor_2 = 1 - (π²/8)(x-1)² - (π/2)(x-1)(y-π/2) - (1/2)(y-π/2)²

# Convert symbolic expressions to fast NumPy functions
f_num    = sp.lambdify((x, y), sin(x*y), 'numpy')
t1_num   = sp.lambdify((x, y), taylor_1_expr, 'numpy')
t2_num   = sp.lambdify((x, y), taylor_2, 'numpy')

# --- Build meshgrid around the expansion point ---
delta = 1.0
Xg = np.linspace(x0_v - delta, x0_v + delta, 120)
Yg = np.linspace(y0_v - delta, y0_v + delta, 120)
Xm, Ym = np.meshgrid(Xg, Yg)

Zf  = f_num(Xm, Ym)
Zt1 = t1_num(Xm, Ym) * np.ones_like(Xm)  # ensure array even if constant
Zt2 = t2_num(Xm, Ym)

# --- Plot ---
fig = plt.figure(figsize=(14, 6))

# ---- Subplot 1: Surface + first-order (tangent plane) ----
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(Xm, Ym, Zf,  alpha=0.55, cmap='viridis',
                 rstride=4, cstride=4, edgecolor='none')
ax1.plot_surface(Xm, Ym, Zt1, alpha=0.45, color='orange',
                 rstride=4, cstride=4, edgecolor='none')
ax1.scatter([x0_v], [y0_v], [1], color='red', s=70, zorder=5,
            label=r'$(1,\,\pi/2)$')
ax1.set_xlabel('$x$'); ax1.set_ylabel('$y$'); ax1.set_zlabel('$z$')
ax1.set_title('1st-order Taylor\n(tangent plane, orange)')
ax1.legend()

# ---- Subplot 2: Surface + second-order (quadratic) ----
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(Xm, Ym, Zf,  alpha=0.55, cmap='viridis',
                 rstride=4, cstride=4, edgecolor='none')
ax2.plot_surface(Xm, Ym, Zt2, alpha=0.45, color='salmon',
                 rstride=4, cstride=4, edgecolor='none')
ax2.scatter([x0_v], [y0_v], [1], color='red', s=70, zorder=5,
            label=r'$(1,\,\pi/2)$')
ax2.set_xlabel('$x$'); ax2.set_ylabel('$y$'); ax2.set_zlabel('$z$')
ax2.set_title('2nd-order Taylor\n(quadratic surface, salmon)')
ax2.legend()

fig.suptitle(r'Taylor approximations of $f(x,y)=\sin(xy)$ at $(1,\,\pi/2)$',
             fontsize=13, y=1.02)
plt.tight_layout()
plt.show()
```

## 4.2 Extrema and the Second Derivative Test

From the notes (§4.2):
1. Find **critical points**: $\nabla f = \mathbf{0}$
2. Classify using the **Hessian** at each critical point:
   - $H_f$ positive definite → local minimum
   - $H_f$ negative definite → local maximum
   - Mixed eigenvalues → saddle point

For two variables, use the **discriminant** $D = f_{xx} f_{yy} - f_{xy}^2$.

```python
# Example from notes: f(x,y,z) = x^2 + y^2 + z^2
f_ext = x**2 + y**2 + z**2
variables_ext = [x, y, z]

# Step 1: Find critical points (∇f = 0)
grad_ext = gradient(f_ext, variables_ext)
critical_pts = solve(grad_ext, variables_ext)
print('Critical points:', critical_pts)

# Step 2: Hessian at the critical point
H_ext = hessian(f_ext, variables_ext)
H_at_cp = H_ext.subs(critical_pts)
print('\nHessian at (0,0,0):')
display(H_at_cp)

# Step 3: Check leading minors (notes: positive definite if all Δ_k > 0)
for k_idx in range(1, len(variables_ext) + 1):
    minor = H_at_cp[:k_idx, :k_idx].det()
    print(f'Δ_{k_idx} = {minor}', '> 0 ✓' if minor > 0 else '')

print('\n→ All leading minors positive → H is positive definite → local MINIMUM at (0,0,0).')
```

```python
# Example from notes: f(x,y) = log(x^2 + y^2 + 1)
f_log = log(x**2 + y**2 + 1)

# Critical points
grad_log = gradient(f_log, [x, y])
cp_log = solve(grad_log, [x, y])
print('Critical point:', cp_log)

# Discriminant D = f_xx * f_yy - f_xy^2 at (0,0)
H_log = hessian(f_log, [x, y]).subs(cp_log)
print('Hessian at (0,0):')
display(H_log)

D_disc = H_log[0, 0] * H_log[1, 1] - H_log[0, 1]**2
print(f'\nD = {D_disc} > 0 and f_xx = {H_log[0,0]} > 0 → local MINIMUM')
```

## 4.3 Lagrange Multipliers

From the notes (§4.3): to find extrema of $f$ subject to $g(\mathbf{x}) = 0$, solve

$$\nabla f = \lambda \nabla g \quad \text{and} \quad g(\mathbf{x}) = 0$$

```python
# Example from notes: minimize f(x,y) = x^2 + y^2 subject to x + y - 1 = 0
f_lag = x**2 + y**2
g_lag = x + y - 1

grad_f_lag = gradient(f_lag, [x, y])
grad_g_lag = gradient(g_lag, [x, y])

# System: ∇f = λ∇g and g = 0
system = [
    grad_f_lag[0] - lam * grad_g_lag[0],  # 2x = λ
    grad_f_lag[1] - lam * grad_g_lag[1],  # 2y = λ
    g_lag,                                   # x + y - 1 = 0
]

solution = solve(system, [x, y, lam])
print('Solution:', solution)

# Value of f at the extremum
if isinstance(solution, dict):
    f_val_lag = f_lag.subs(solution)
else:
    f_val_lag = f_lag.subs([(x, solution[0][0]), (y, solution[0][1])])
print('f at extremum =', f_val_lag)  # 1/2
```

```python
# Multiple constraints (Example from notes):
# Minimize f(x,y,z) = x^2+y^2+z^2 subject to g1: x+y+z=1 and g2: x-y=0

f_mc = x**2 + y**2 + z**2
g1_mc = x + y + z - 1
g2_mc = x - y

lam1, lam2 = symbols('lambda_1 lambda_2', real=True)

grad_f_mc = gradient(f_mc, [x, y, z])
grad_g1 = gradient(g1_mc, [x, y, z])
grad_g2 = gradient(g2_mc, [x, y, z])

# ∇f = λ1·∇g1 + λ2·∇g2
system_mc = [
    grad_f_mc[0] - lam1*grad_g1[0] - lam2*grad_g2[0],
    grad_f_mc[1] - lam1*grad_g1[1] - lam2*grad_g2[1],
    grad_f_mc[2] - lam1*grad_g1[2] - lam2*grad_g2[2],
    g1_mc,
    g2_mc,
]

sol_mc = solve(system_mc, [x, y, z, lam1, lam2])
print('Solution:', sol_mc)
print('f at extremum =', f_mc.subs(sol_mc))  # 1/3
```

## Exercise 2 — Differentiation and Classification of Critical Points

*(Relates to Notes §3.1, §3.4, §4.2)*

Consider the function:
$$f(x, y) = x^3 - 3xy + y^3$$

1. Compute the **gradient** $\nabla f$.
2. Find all **critical points** (solve $\nabla f = \mathbf{0}$).
3. Compute the **Hessian matrix** $H_f$.
4. Use the **second derivative test** (discriminant $D = f_{xx}f_{yy} - f_{xy}^2$) to classify each critical point.
5. Compute the **directional derivative** of $f$ at $(1, 1)$ in the direction $\mathbf{u} = \left(\frac{3}{5}, \frac{4}{5}\right)$.

```python
# Exercise 2 — Skeleton

import sympy as sp
from sympy import (
    symbols, sqrt, cos, sin, tan, atan2, acos, pi, exp, log,
    Matrix, Rational, Function, simplify, trigsimp, latex, Abs,
    diff, solve, Eq, oo, limit, Piecewise,
)
x, y = symbols('x y')
f_ex2 = x**3 - 3*x*y + y**3

# 1. Gradient
grad_ex2 = Matrix([f_ex2]).jacobian([x, y]).T   # TODO: use the gradient() helper defined above
print('∇f =');  
display(grad_ex2)

# 2. Critical points
cps = solve(grad_ex2, [x, y])
print('Critical points:', cps)

# 3. Hessian
H_ex2 = grad_ex2.jacobian([x, y])     # TODO: use the hessian() helper
print('Hessian H_f ='); 
display(H_ex2)

# 4. Second derivative test at each critical point
# For each cp, compute D = H[0,0]*H[1,1] - H[0,1]^2
# and check sign of H[0,0] to classify
print("Classification of critical points: ")
for cp in cps:
    if all(val.is_real for val in cp):
        H_val = H_ex2.subs({x: cp[0], y: cp[1]}) # TODO: substitute the critical point
        D_val = H_val.det()       # TODO: compute discriminant
        fxx = H_val[0,0]               # TODO: H_val[0,0]
        if D_val > 0 and fxx > 0:
            kind = 'min'
        elif  D_val > 0 and fxx < 0:
            kind = 'max'
        elif D_val < 0:
            kind = 'saddle'
        else:
            kind = 'inconclusive'
    
    print(f'At {cp}: D = {D_val}, f_xx = {fxx} → {kind}')

# 5. Directional derivative at (1,1) in direction u = (3/5, 4/5)
u_ex2 = Matrix([Rational(3, 5), Rational(4, 5)])
grad_at_11 = grad_ex2.subs({x : 1, y : 1})  # TODO: evaluate gradient at (1,1)
D_u_ex2 = grad_at_11.dot(u_ex2)    # TODO: dot product
print('D_u f(1,1) =', D_u_ex2)
```

## Exercise 3 — Lagrange Multipliers and the Jacobian

*(Relates to Notes §3.5, §4.3)*

**Part A — Constrained optimization:**  
Find the extrema of $f(x, y, z) = xyz$ subject to the constraint $g(x, y, z) = x^2 + y^2 + z^2 - 3 = 0$.

**Part B — Jacobian and chain rule:**  
Let $\mathbf{f}(u, v) = (u^2 - v^2, \; 2uv)$ and $\mathbf{g}(t) = (\cos t, \; \sin t)$.  
Compute the Jacobian $D\mathbf{f}$, the Jacobian $D\mathbf{g}$, and verify the chain rule $D(\mathbf{f}\circ\mathbf{g})(t) = D\mathbf{f}(\mathbf{g}(t))\cdot D\mathbf{g}(t)$.

```python
# Exercise 3A — Lagrange Multipliers Skeleton
import sympy as sp
from sympy import (
    symbols, sqrt, cos, sin, tan, atan2, acos, pi, exp, log,
    Matrix, Rational, Function, simplify, trigsimp, latex, Abs,
    diff, solve, Eq, oo, limit, Piecewise,
)

#I define the symbols
x, y, z, lam = symbols('x y z lam')

f_ex3 = x * y * z
g_ex3 = x**2 + y**2 + z**2 - 3

# Build the system ∇f = λ∇g and g = 0
grad_f_ex3 = Matrix([f_ex3]).jacobian([x, y, z]).T  # TODO
grad_g_ex3 =  Matrix([g_ex3]).jacobian([x, y, z]).T   # TODO

system_ex3 = [
    grad_f_ex3[0] - lam * grad_g_ex3[0],
    grad_f_ex3[1] - lam * grad_g_ex3[1],
    grad_f_ex3[2] - lam * grad_g_ex3[2],
    g_ex3
]
# TODO: 3 equations from ∇f = λ∇g
#     ...,  # TODO: constraint g = 0
# ]
sol_ex3 = solve(system_ex3, [x, y, z, lam])
print('Solutions:', sol_ex3)
print('Solutions found', len(sol_ex3))

# Evaluate f at each solution to find max and min
val_f = []
for s in sol_ex3:
    x_val, y_val, z_val = s[0], s[1],s[2]
    val_r = f_ex3.subs({x: x_val, y: y_val, z: z_val})
    val_f.append(val_r)
    print(f'Point {s[:3]} -> f = {val_r}')
    print(f'Maximum value: {max(val_f)}')
    print(f'Minimum value: {min(val_f)}')
```

```python
# Exercise 3B — Jacobian and Chain Rule Skeleton
import sympy as sp
from sympy import (
    symbols, sqrt, cos, sin, tan, atan2, acos, pi, exp, log,
    Matrix, Rational, Function, simplify, trigsimp, latex, Abs,
    diff, solve, Eq, oo, limit, Piecewise,
)

u, v = symbols('u v', real=True)
t = symbols('t', real = True)

f_ex3b = Matrix([u**2 - v**2, 2*u*v])
g_ex3b = Matrix([cos(t), sin(t)])

# Compute Df (2×2 Jacobian)
Df_ex3b = f_ex3b.jacobian([u, v])  # TODO: f_ex3b.jacobian([u, v])
print('Df =')
display(Df_ex3b)

# Compute Dg (2×1 Jacobian)
Dg_ex3b = g_ex3b.jacobian([t])  # TODO: g_ex3b.jacobian([t])
print('Dg =')
display(Dg_ex3b)

# Chain rule: D(f∘g) = Df(g(t)) · Dg(t)
Df_at_g = Df_ex3b.subs({u: cos(t), v: sin(t)})  # TODO: substitute u=cos(t), v=sin(t)
D_comp = Df_at_g * Dg_ex3b
print('D(f∘g)(t) via chain rule =')
display(simplify(D_comp))

# Verify by direct computation
f_composed = f_ex3b.subs([(u, cos(t)), (v, sin(t))])
D_direct = f_composed.jacobian([t])
print('D(f∘g)(t) direct =')
display(simplify(D_direct))
print('Both agree:', simplify(D_comp - D_direct) == sp.zeros(2, 1))
```
